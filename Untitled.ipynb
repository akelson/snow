```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def uniformDistPtsTriangle(n):
    x = np.random.rand(2, n)
    x[1, :] = np.sqrt(x[1, :])
    x[0, :] = x[1, :] * (x[0, :] - .5)
    return x
```

```python
x = uniformDistPtsTriangle(500)
```

```python
plt.plot(x[0, :], x[1, :], '.')
```
